```python
userinput = input("enter the symbol")
a = int(input("enter a value"))
b = int(input("enter  b value"))

if(userinput == "+"):
    d = a+b
    print("addition is d",d)
    if(d>60):
        print("nothing")
elif(userinput == "*"):
    d = a*b
    print("multiplication is",d)
elif(userinput =="-"):
    d = a-b
    print("multiplication is",d)
elif(userinput =="/"):
    d = a/b
    print("multiplication is",d)
else:
    print("no operation is given")
```
