```python
from sklearn.model_selection import train_test_split                #previously cros_validation was used in sklearn
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
x_train = sc1.fit_transform(x_train)
x_test = sc1.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

# if you want to print colums in x

```python
x_train.shape
```

adding layers  belowis the input layer

```python
model.add(Dense(input_dim=11,init="random_uniform",activation='relu',output_dim=6))
```

# adding hidden layer

```python
model.add(Dense(init="random_uniform",activation='relu',output_dim=6))
```

# adding output layer

```python
model.add(Dense(output_dim=1,init="random_uniform",activation='sigmoid'))
```

# if you use sigmoid the output will be from 0 to 1 then we will write threshold

# for calculating error we will use binary cross entropy or Log loss

# compile the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# standard scalar  or future scaling mainly for classification

```python
model.fit(x_train,y_train,epochs=2,batch_size=32)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.min()
```

```python
y_pred = (y_pred>0.5)
```
